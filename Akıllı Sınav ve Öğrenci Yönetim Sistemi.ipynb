```python
# Sınıf (class) kullanmadığımız için verileri liste içinde sözlük (dict) olarak tutuyoruz.
ogrenciler = []
MAX_OGRENCI = 30


def durum_hesapla(vize, final, devamsizlik):
    """
    Sistemin karar mekanizması.
    Hesaplamaları yapıp sonucu döndürür (Local Scope).
    """
    if devamsizlik > 10:
        return 0, "KALDI (Devamsız)"

    # Ortalama hesaplama mantığı
    ortalama = (vize * 0.4) + (final * 0.6)
    ortalama = round(ortalama, 2)

    if ortalama >= 60:
        durum = "GEÇTİ"
    elif 50 <= ortalama < 60:
        durum = "BÜTÜNLEME"
    else:
        durum = "KALDI"

    return ortalama, durum


def ogrenci_ekle():
    if len(ogrenciler) >= MAX_OGRENCI:
        print("\n[!] HATA: Kontenjan dolu!")
        return

    o_id = input("Öğrenci ID: ")

    # ID Tekillik Kontrolü
    for ogrenci in ogrenciler:
        if ogrenci["id"] == o_id:
            print("\n[!] HATA: Bu ID zaten sistemde mevcut!")
            return

    ad = input("Öğrenci Adı Soyadı: ")

    # Yeni öğrenci taslağı
    yeni_ogrenci = {
        "id": o_id,
        "ad": ad,
        "vize": 0,
        "final": 0,
        "devamsizlik": 0,
        "ortalama": 0,
        "durum": "Not Girilmedi"
    }

    ogrenciler.append(yeni_ogrenci)
    print(f"\n[+] {ad} başarıyla sisteme kaydedildi.")


def not_gir():
    o_id = input("Not girişi yapılacak öğrenci ID: ")

    for ogrenci in ogrenciler:
        if ogrenci["id"] == o_id:
            print(f"\n--- {ogrenci['ad']} Not Girişi ---")

            # Veri Doğrulama (Input Validation)
            while True:
                try:
                    v = float(input("Vize (0-100): "))
                    f = float(input("Final (0-100): "))
                    d = int(input("Devamsızlık: "))

                    if 0 <= v <= 100 and 0 <= f <= 100 and d >= 0:
                        break
                    else:
                        print("[!] HATA: Geçersiz değerler! Tekrar deneyin.")
                except ValueError:
                    print("[!] HATA: Lütfen sayısal bir değer girin.")

            # Hesaplama fonksiyonundan veriyi al ve güncelle
            ort, sonuc = durum_hesapla(v, f, d)

            ogrenci.update({
                "vize": v,
                "final": f,
                "devamsizlik": d,
                "ortalama": ort,
                "durum": sonuc
            })
            print("\n[✓] Notlar ve başarı durumu güncellendi.")
            return

    print("\n[!] HATA: Öğrenci bulunamadı.")


def ogrenci_listele():
    if not ogrenciler:
        print("\n[!] Listelenecek öğrenci bulunmuyor.")
        return

    print("\n" + "=" * 75)
    print(f"{'ID':<10} {'AD SOYAD':<20} {'VİZE':<8} {'FİNAL':<8} {'ORT.':<8} {'DURUM'}")
    print("-" * 75)

    for og in ogrenciler:
        print(f"{og['id']:<10} {og['ad']:<20} {og['vize']:<8} {og['final']:<8} {og['ortalama']:<8} {og['durum']}")
    print("=" * 75)


def butunleme_islemi():
    o_id = input("Bütünleme sınavına girecek öğrenci ID: ")

    for ogrenci in ogrenciler:
        if ogrenci["id"] == o_id:
            if ogrenci["durum"] != "BÜTÜNLEME":
                print(f"\n[!] Bu öğrenci bütünlemeye giremez! Mevcut Durum: {ogrenci['durum']}")
                return

            while True:
                try:
                    yeni_but = float(input("Bütünleme Notu (Final yerine geçer): "))
                    if 0 <= yeni_but <= 100:
                        break
                    print("[!] 0-100 arası girin.")
                except ValueError:
                    print("[!] Sayı girin.")

            # Bütünleme notu finali ezer
            ogrenci["final"] = yeni_but
            ort, sonuc = durum_hesapla(ogrenci["vize"], ogrenci["final"], ogrenci["devamsizlik"])

            ogrenci["ortalama"] = ort
            ogrenci["durum"] = sonuc
            print(f"\n[✓] Bütünleme işlendi. Yeni Durum: {sonuc}")
            return

    print("\n[!] Öğrenci bulunamadı.")


def analiz_yap():
    if not ogrenciler:
        print("\n[!] Analiz için veri yok.")
        return

    toplam = sum(o["ortalama"] for o in ogrenciler)
    gecenler = [o for o in ogrenciler if o["durum"] == "GEÇTİ"]
    kalanlar = [o for o in ogrenciler if o["durum"] in ("KALDI", "KALDI (Devamsız)")]

    # Algoritma: En yüksek ve en düşük notu bulma
    en_yuksek = max(o["ortalama"] for o in ogrenciler)
    en_dusuk  = min(o["ortalama"] for o in ogrenciler)

    print("\n" + "--- SINIF ANALİZ RAPORU ---".center(40))
    print(f"Öğrenci Sayısı    : {len(ogrenciler)}")
    print(f"Sınıf Ortalaması  : {toplam / len(ogrenciler):.2f}")
    print(f"Geçen Sayısı      : {len(gecenler)}")
    print(f"Kalan Sayısı      : {len(kalanlar)}")
    print(f"Başarı Oranı      : %{(len(gecenler) / len(ogrenciler)) * 100:.1f}")
    print(f"En Yüksek Ortalama: {en_yuksek}")
    print(f"En Düşük Ortalama : {en_dusuk}")
    print("-" * 40)


def ana_menu():
    while True:
        print("\n--- AKILLI ÖĞRENCİ YÖNETİM SİSTEMİ ---")
        print("1- Öğrenci Ekle")
        print("2- Not/Devamsızlık Gir")
        print("3- Listele")
        print("4- Analiz Yap")
        print("5- Bütünleme Sınavı")
        print("0- Çıkış")

        secim = input("\nSeçiminiz: ")

        if secim == "1":
            ogrenci_ekle()
        elif secim == "2":
            not_gir()
        elif secim == "3":
            ogrenci_listele()
        elif secim == "4":
            analiz_yap()
        elif secim == "5":
            butunleme_islemi()
        elif secim == "0":
            print("Sistem kapatılıyor. İyi günler!")
            break
        else:
            print("[!] Geçersiz seçim!")


# Programın Giriş Noktası
if __name__ == "__main__":
    ana_menu()
```
